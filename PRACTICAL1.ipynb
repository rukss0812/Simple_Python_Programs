```
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Define a number for which to compute the factorial
num_to_factorialize = 5
fact_result_new = factorial(num_to_factorialize)
print(f"The factorial of {num_to_factorialize} is {fact_result_new}")
```

```
print(f"Multiplication Table for {num_to_factorialize}:")
for i in range(1, 11):
    print(f"{num_to_factorialize} x {i} = {num_to_factorialize * i}")
```

```
if num_to_factorialize % 2 == 0:
    result = "Even"
else:
    result = "Odd"
print(f"The number {num_to_factorialize} is {result}.")
```

```
def is_palindrome(n):
    # Convert number to string to easily reverse and compare
    s = str(n)
    return s == s[::-1]

# Using the 'num_to_factorialize' variable from the kernel state
if is_palindrome(num_to_factorialize):
    print(f"The number {num_to_factorialize} is a palindrome.")
else:
    print(f"The number {num_to_factorialize} is not a palindrome.")
```

# Task
Develop an expert system for animal identification based on 'if-then' rules. Implement a knowledge base for these rules, create an inference engine to deduce conclusions from given facts, and test the system with various animal characteristics to verify its decision-making.

## Define Problem Domain and Rules

### Subtask:
Define a specific problem domain for the expert system and create a set of 'if-then' rules that represent the expert knowledge for this domain.


### Problem Domain: Animal Identification

The goal of this expert system is to identify an animal based on a set of observed characteristics. We will define a set of 'if-then' rules to classify animals into categories such as Mammal, Bird, Reptile, Fish, Amphibian, or Insect.

#### Animal Characteristics and Rules:

For each animal type, we will establish key distinguishing characteristics:

*   **Mammal:**
    *   `Has fur`
    *   `Gives milk`
    *   `IF Has fur AND Gives milk THEN Animal is a Mammal`

*   **Bird:**
    *   `Has feathers`
    *   `Can fly`
    *   `IF Has feathers AND Can fly THEN Animal is a Bird`

*   **Reptile:**
    *   `Lays eggs`
    *   `Has scales`
    *   `IF Lays eggs AND Has scales THEN Animal is a Reptile`

*   **Fish:**
    *   `Lays eggs`
    *   `Lives in water`
    *   `Has gills`
    *   `IF Lays eggs AND Lives in water AND Has gills THEN Animal is a Fish`

*   **Amphibian:**
    *   `Has moist skin`
    *   `Lays eggs`
    *   `Lives in water and land`
    *   `IF Has moist skin AND Lays eggs AND Lives in water and land THEN Animal is an Amphibian`

*   **Insect:**
    *   `Has exoskeleton`
    *   `Segmented body`
    *   `Six legs`
    *   `IF Has exoskeleton AND Segmented body AND Six legs THEN Animal is an Insect`

#### Conclusion for Unknown:

If the given facts do not match any of the defined rules, the conclusion will be:

*   `No conclusion can be drawn from the given facts.`

## Implement Knowledge Base

### Subtask:
Implement the defined rules in a structured format to serve as the expert system's knowledge base.


**Reasoning**:
First, I'll define variables to store the consequence strings for clarity and reusability, including a variable for an unknown conclusion. Then, I'll create the `rules` list, with each element being a dictionary containing 'conditions' and 'consequence' for each animal type, as specified in the instructions. This directly implements the knowledge base as a structured Python list.



```
conclusion_bird = 'Animal is a Bird'
conclusion_mammal = 'Animal is a Mammal'
conclusion_reptile = 'Animal is a Reptile'
conclusion_fish = 'Animal is a Fish'
conclusion_amphibian = 'Animal is an Amphibian'
conclusion_insect = 'Animal is an Insect'
conclusion_unknown = 'No conclusion can be drawn from the given facts.'

rules = [
    {
        'conditions': ['Has feathers', 'Can fly'],
        'consequence': conclusion_bird
    },
    {
        'conditions': ['Has fur', 'Gives milk'],
        'consequence': conclusion_mammal
    },
    {
        'conditions': ['Lays eggs', 'Has scales'],
        'consequence': conclusion_reptile
    },
    {
        'conditions': ['Lays eggs', 'Lives in water', 'Has gills'],
        'consequence': conclusion_fish
    },
    {
        'conditions': ['Has moist skin', 'Lays eggs', 'Lives in water and land'],
        'consequence': conclusion_amphibian
    },
    {
        'conditions': ['Has exoskeleton', 'Segmented body', 'Six legs'],
        'consequence': conclusion_insect
    }
]

print("Knowledge base (rules) and conclusion variables defined.")
```

## Develop Inference Engine

### Subtask:
Create a Python function that acts as the inference engine, taking a set of facts (user inputs) and applying the rules from the knowledge base to deduce a conclusion or decision using a simple forward-chaining mechanism.


**Reasoning**:
I will create the `infer_animal_type` function to implement the forward-chaining inference engine as described in the instructions. This function will check provided facts against the predefined rules and return a conclusion.



```
def infer_animal_type(facts):
    """
    Inference engine to deduce animal type based on given facts using forward-chaining.

    Args:
        facts (list): A list of characteristics observed for an animal.

    Returns:
        str: The deduced animal type or 'No conclusion can be drawn from the given facts.'
    """
    for rule in rules:
        # Check if all conditions for the current rule are present in the provided facts
        if all(condition in facts for condition in rule['conditions']):
            return rule['consequence'] # If all conditions met, return the consequence

    # If no rule matches, return the unknown conclusion
    return conclusion_unknown

print("Inference engine function 'infer_animal_type' defined.")
```
